```python
# Get whl file from MK repo
import requests
import subprocess

github_url = "https://api.github.com/repos/m-kovalsky/Fabric/contents/Direct%20Lake%20Migration"
response = requests.get(github_url)
for item in response.json():
    if item['type'] == 'file' and item['name'].endswith('.whl'):
        whlFile = item['download_url']
        subprocess.run(["pip", "install", whlFile])
```
